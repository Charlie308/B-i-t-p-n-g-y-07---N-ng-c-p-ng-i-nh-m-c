```python
import math
import turtle
t = turtle.Turtle()
turtle.bgcolor("pink")
t.speed(0)
t.pensize(2)
l = 300 ## chiều dài ngôi nhà
h = 200 ## chiều cao nhà
r = 150 ## chiều dài mái nhà
ld = 60 ## chiều dài cửa chính
hd = 100 ## chiều cao cửa chính
w = 40 ## cửa sổ nhà

t.penup()
t.goto(-200,-200)
t.pendown()
t.pencolor("red")

##vẽ nhà
t.fillcolor("blue")
t.begin_fill()
for i in range(2):
    t.fd(l)
    t.lt(90)
    t.fd(h)
    t.lt(90)
t.end_fill()

t.lt(90)
t.fd(h)

t.fillcolor("orange")
t.begin_fill()
t.rt(45)
t.fd(r)
t.rt(90)
t.fd(r)
t.end_fill()

t.bk(r/3)

t.fillcolor("brown")
t.begin_fill()
t.lt(135)
t.fd(r/2)
t.lt(90)
t.fd(r/4)
t.lt(90)
t.fd(r/4.4)
t.end_fill()

t.penup()
t.goto(-130,-200)
t.lt(90)
t.pendown()
t.fillcolor("yellow")
t.begin_fill()
for i in range(2):
    t.fd(ld)
    t.lt(90)
    t.fd(hd)
    t.lt(90)
t.end_fill()
    
t.penup()
t.goto(0,-100)
t.pendown()
t.fillcolor("yellow")
t.begin_fill()
for i in range(4):
    t.fd(w)
    t.lt(90)
t.end_fill()

##vẽ cây xanh
##biến số gốc cây
x = 30
y = 70
## biến số lá cây
m = 90


t.penup()
t.goto(-300,-250)
t.pendown()
t.fillcolor("brown")
t.begin_fill()
for i in range(2):
    t.fd(x)
    t.lt(90)
    t.fd(y)
    t.lt(90)
t.end_fill()

t.fd(x/2)

for i in range(3):
    t.penup()
    t.lt(90)
    t.fd(y)
    t.rt(90)
    t.bk(m/2)
    t.pendown()

    t.fillcolor("green")
    t.begin_fill()
    for i in range(3):
        t.fd(m)
        t.lt(120)
    t.end_fill()
    t.fd(m/2)
    
x *= 1.2
y *= 1.2
m *= 1.2
    
t.penup()
t.goto(200,-100)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
for i in range(2):
    t.fd(x)
    t.lt(90)
    t.fd(y)
    t.lt(90)
t.end_fill()

t.fd(x/2)

for i in range(3):
    t.penup()
    t.lt(90)
    t.fd(y)
    t.rt(90)
    t.bk(m/2)
    t.pendown()

    
    t.fillcolor("green")
    t.begin_fill()
    for i in range(3):
        t.fd(m)
        t.lt(120)
    t.fd(m/2)
    t.end_fill()
```
